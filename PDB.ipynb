```python
from pdbfixer import PDBFixer
from openmm.app import PDBFile
from openmm.app import Modeller
from openmm.app import ForceField
from openmm.app import element
from openmm.app import Simulation
from openmm.app import PME, HBonds, NoCutoff
from openmm.unit import *
from sys import stdout
import openmm

# Cargar estructura
fixer = PDBFixer(pdbid='4UNN')

# Mantener solo cadena A
fixer.chain_id = 'A'
fixer.removeChains([i for i, c in enumerate(fixer.topology.chains()) if c.id != 'A'])

# Eliminar heteroÃ¡tomos
fixer.removeHeterogens(keepWater=False)

# Completar residuos faltantes
fixer.findMissingResidues()
fixer.findMissingAtoms()
fixer.addMissingAtoms()

# AÃ±adir hidrÃ³genos para pH 7
fixer.addMissingHydrogens(pH=7.0)

# Guardar estructura limpia
PDBFile.writeFile(fixer.topology, fixer.positions, open('receptor_clean.pdb', 'w'))
```
